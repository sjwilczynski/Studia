```python
k = 100
d = 10
def macierz_obrotu(alpha):
    return np.matrix([[np.cos(alpha), -np.sin(alpha)], [np.sin(alpha), np.cos(alpha)]])
alpha = 2*np.pi/k
radius = np.sqrt(d**2/(2*(1 - np.cos(alpha))))
punkt = np.matrix([[radius],[0]])
for i in range(0,k):
    s1 = punkt[0][0]
    s2 = punkt[1][0]     
    a = np.random.normal(s1,1,10)
    b = np.random.normal(s2,1,10)
    #scatter(a,b,c = ??) jak działa dobieranie kolorów???
    plot(a,b,'o')
    punkt = macierz_obrotu(alpha)*punkt
show()
    
        
        
```
