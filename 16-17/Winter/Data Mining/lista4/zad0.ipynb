```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score

from sklearn.cluster import KMeans, MiniBatchKMeans, Birch, DBSCAN
```

```python
import matplotlib.colors as colors
from itertools import cycle

def plotClustering(X, plt_labels):
    plt_colors = cycle(colors.cnames.keys())

    plt_K = np.unique(plt_labels).size

    for k in xrange(plt_K):
        color = plt_colors.next()
        mask = (plt_labels == k)
        plt.plot(X[plt_labels == -1, 0], X[plt_labels == -1, 1], linestyle = 'None', markerfacecolor = 'black', marker = 'o')
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor=color, marker='o')

    plt.show()
```

```python
def distance(x,y):
    return np.sqrt(np.sum(x**2,axis = 1,keepdims = True) + np.sum(y**2,axis = 1,keepdims = True).T - 2*np.dot(x,y.T))

def dunn_index(X,labels):
    k = (np.unique(labels) >= 0).sum()
    intra = np.max([np.max(distance(X[labels == i], X[labels == i])) for i in range(k)])
    inter = np.min([np.min(distance(X[labels == i], X[labels == j])) for i in range(k) for j in range(i+1,k)])
    return inter/intra

def david_bouldin_index(X, labels):
    k = (np.unique(labels) >= 0).sum()
    centroids = [X[labels==i].mean(0) for i in range(k)]
    avg_c_dists = [np.sqrt(((X[labels==i] - centroids[i])**2).sum(1)).mean() for i in range(k)]
    
    ratio = lambda x, y: (avg_c_dists[x] + avg_c_dists[y])/np.sqrt(((centroids[x]-centroids[y])**2).sum())
    
    return np.mean([max([ratio(i, j) if i != j else 0 for j in range(k)]) for i in range(k)])
    
```

```python
from sklearn import datasets
plt.figure(figsize=((15,10)))
centers_ = [[1, 1], [3, 3],[5,1]]
X, labels = datasets.make_blobs(n_samples=3000, n_features=2, centers=centers_, cluster_std=0.5)

plotClustering(X, labels)
```

```python
k_means = KMeans(init='k-means++', n_clusters=3, n_init=100)
k_means.fit(X)
plt.figure(figsize=((15,10)))

print "Silhoute: ",silhouette_score(X, k_means.labels_) #im wiekszy tym lepszy
print "Dunn: ", dunn_index(X, k_means.labels_)# im wiekszy tym lepszy
print "David_Bouldin_: ", david_bouldin_index(X, k_means.labels_) #im mniejszy tym lepszy
plotClustering(X, k_means.labels_)
print k_means.inertia_
```

```python
birch = Birch(threshold=0.85, n_clusters=None)
birch.fit(X)
plt.figure(figsize=((15,10)))

print "Silhoute: ",silhouette_score(X, birch.labels_) #im wiekszy tym lepszy
print "Dunn: ", dunn_index(X,birch.labels_)# im wiekszy tym lepszy
print "David_Bouldin_: ", david_bouldin_index(X,birch.labels_) #im mniejszy tym lepszy
plotClustering(X, birch.labels_)
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=25)
dbscan.fit(X)
plt.figure(figsize=((15,10)))

print "Silhoute: ",silhouette_score(X, dbscan.labels_) #im wiekszy tym lepszy
print "Dunn: ", dunn_index(X,dbscan.labels_)# im wiekszy tym lepszy
print "David_Bouldin_: ", david_bouldin_index(X,dbscan.labels_) #im mniejszy tym lepszy
plotClustering(X, dbscan.labels_)
```
